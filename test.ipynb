```python
def company_with_most_worst_car(d):
    print("\n\n\n\nsearching")
    company_name=""
    worst_car_count=0
    for name,count in d.items():
        if len(count)>worst_car_count:
            company_name=name
            worst_car_count=len(count)
    return company_name



def main():

    car_dict={}
    file_path="cars.txt"
    print("Cars:")
    print("Year     Model")
    with open(file_path,'r') as read_file:
        for line in read_file:
            line=line.strip()
            if len(line)==0:
                continue
            else:
                line.strip('\'()')
                line = line.split()
                year = line[-1].strip('\'()')

        
                if len(line)==2:
                    if car_dict.get(line[0])==None:
                        cars = []
                        manu= line[1]
#                         year= ' '.join(line.split()[1:])
                        T = line[0],year
                        print(T)
                        
                        cars.append(T)
                        car_dict[line[0]]= cars
                    else:
                        cars = car_dict.get(line[0])
                        T = line[0],year
                        
                        cars.append(T)
                        print(T)
        
                elif len(line)==1:
                    if car_dict.get(line[0])==None:
                        cars = []
                        T = line[0], ""
                        cars.append(T)
                        print(T)
                        
                        car_dict[line[0]]= cars
                    else:
                        cars = car_dict.get(line[0])

                        T = line[0],""
                        cars.append(T)
                        print(T)
                else:
                    if car_dict.get(line[0])==None:
                        cars = []
                        T = line[0]," ".join(line[1:])
                        cars.append(T)
                        print(T)
                        car_dict[line[0]]= cars
                    else:
                        cars = car_dict.get(line[0])
                        T = line[0]," ".join(line[1:])
                        cars.append(T)
                        print(T)

    print("Worst manufacturer year:",company_with_most_worst_car(car_dict))



main()
```
